#Forecasting Replenishment
This code runs several different methods and compares their accuracy

*Using just a minimal data-set (daily sales quantities @ one store of 1 item)*

```python
#importing some useful libraries

%matplotlib inline
import datetime as dt
from dateutil import parser
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from dateutil.relativedelta import relativedelta

from matplotlib.dates import YearLocator, MonthLocator, DateFormatter

import seaborn as sns #pretty graphs
import scipy #for statistics (like r^2)
from bisect import bisect_left#i used this just for making a 'find closest time' search
```

```python
#opening and formatting data

#datafile = open('daily_HistoricalSales_1item1store - Sheet1.csv', 'r')
datafile = open('us_diesel_sales_price', 'r')

raw_data = []
for row in datafile:
    temp = row.strip().split(',')
    #raw_data.append([temp[0].split(' ')[0], temp[4]])
    raw_data.append([temp[0], temp[1]])
raw_data = raw_data[1:]
    
formatted_data = []
for d in raw_data:
    formatted_data.append((parser.parse(d[0]),float(d[1])));

#final data set for 'real values'    
dates = [q[0] for q in formatted_data]
prices = [q[1] for q in formatted_data]
```

###Next we will do some simple math to calculate some forecasts


```python
''' helper funcs '''

# returns the index of the datetime which is closest to exactly a year before input date
# input: list of dts, this dt
def ret_dt_idx(dts, dt):
    dt -= relativedelta(years=1)
    min_idx = 0
    min_delta = abs(dts[0] - dts[-1])
    for idx, i in enumerate(dts):
        delta = abs(dt - i)
        if delta < min_delta:
            min_delta = delta
            min_idx = idx  
    return min_idx


''' forecasting methods ''' 

#just use this week as our prediction
prices1 = []
for i in xrange(len(prices)):
	if i>4: 
		x = (prices[i-3])
		prices1.append(x)
	else:
		prices1.append(prices[i])

#averaging this week and last week (three weeks out)
prices2 = []
for i in xrange(len(prices)):
	if i>5: 
		x = (prices[i-3] + prices[i-4])/2
		prices2.append(x)
	else:
		prices2.append(prices[i])

#max of this week and TYLY
prices3 = []
for i in xrange(len(prices)):
    which_idx = ret_dt_idx(dates,dates[i])
    x = max(prices[i-3],prices[which_idx])
    prices3.append(x)

#max of this week and last week
prices4 = []  
for i in xrange(len(prices)):
    if i>5: 
        x = max(prices[i-4],prices[i-3])
        prices4.append(x)
    else:
        prices4.append(prices[i])
        
        
```

###Now let's make two plots

1. The forecasts on top of the actual data
2. And the differences between the forecasts and the actual data

```python
# this figure will contain two plots
fig = plt.figure()
fig.set_size_inches(18.5,13.5)
fig.suptitle('Comparing Forecast Models', fontsize=20)

# range to show
start = 450
end = -1

# plot 1
ax1 = fig.add_subplot(211)
ax1.xaxis_date()
ax1.plot(dates[start:end], prices[start:end], 'k', label='Actual')
ax1.plot(dates[start:end], prices1[start:end], label='using TW')
ax1.plot(dates[start:end], prices2[start:end], label='Avg of TW & LW')
ax1.plot(dates[start:end], prices3[start:end], label='LY mod')
ax1.plot(dates[start:end], prices4[start:end], label='Max of TW & LW')
ax1.legend(loc='upper left')

# calculate diffs
diffs1 = [j-i for (i,j) in zip(prices,prices1)]
diffs2 = [j-i for (i,j) in zip(prices,prices2)]
diffs3 = [j-i for (i,j) in zip(prices,prices3)]
diffs4 = [j-i for (i,j) in zip(prices,prices4)]

# plot 2
ax2 = fig.add_subplot(212)
ax2.xaxis_date()
ax2.plot(dates[start:end], diffs1[start:end])
ax2.fill_between(dates[start:end], 0, diffs1[start:end], facecolor='blue', alpha=0.2)
ax2.plot(dates[start:end], diffs2[start:end])
ax2.fill_between(dates[start:end], 0, diffs2[start:end], facecolor='green', alpha=0.2)
ax2.plot(dates[start:end], diffs3[start:end])
ax2.fill_between(dates[start:end], 0, diffs3[start:end], facecolor='red', alpha=0.2)
ax2.plot(dates[start:end], diffs3[start:end])
ax2.fill_between(dates[start:end], 0, diffs3[start:end], facecolor='purple', alpha=0.2)

fig.autofmt_xdate()
```

### A simpler metric...

And let's do a little math to reduce these graphs to some numbers which can be more easily compared.


```python
def r2(data, predictions):
    return scipy.stats.pearsonr(data, predictions)[0]**2

def mape(f,a):
    x = 0.0
    for i,j in zip(f,a):
        x += abs(i-j)/j
    return (1 - x/len(f)) * 100



print 'MAPE values...'
print 'method 1:', mape(prices[start:end], prices1[start:end])
print 'method 2:', mape(prices[start:end], prices2[start:end])
print 'method 3:', mape(prices[start:end], prices3[start:end])
print 'method 4:', mape(prices[start:end], prices4[start:end])
```
